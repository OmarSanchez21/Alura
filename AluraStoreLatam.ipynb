### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
tienda2.head()
tienda3.head()
tienda4.head()
pd.options.display.float_format = '{:.2f}'.format
tienda
```

#1. Análisis de facturación



```python
ingreso_tienda1 = tienda["Precio"].sum()
ingreso_tienda2 = tienda2["Precio"].sum()
ingreso_tienda3 = tienda3["Precio"].sum()
ingreso_tienda4 = tienda4["Precio"].sum()

ingresos = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Ingresos": [
        ingreso_tienda1,
        ingreso_tienda2,
        ingreso_tienda3,
        ingreso_tienda4
    ]
})
ingresos
```

# 2. Ventas por categoría

```python
agrupacion = tienda.groupby("Categoría del Producto").size().sort_values(ascending=False)
agrupacion
```

```python
agrupacion2 = tienda2.groupby("Categoría del Producto").size().sort_values(ascending=False)
agrupacion2
```

```python
agrupacion3 = tienda3.groupby("Categoría del Producto").size().sort_values(ascending=False)
agrupacion3
```

```python
agrupacion4 = tienda4.groupby("Categoría del Producto").size().sort_values(ascending=False)
agrupacion4
```

# 3. Calificación promedio de la tienda


```python
promedio_tienda1 = tienda["Calificación"].mean()
promedio_tienda2 = tienda2["Calificación"].mean()
promedio_tienda3 = tienda3["Calificación"].mean()
promedio_tienda4 = tienda4["Calificación"].mean()

promedios = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Calificación Promedio": [
        promedio_tienda1,
        promedio_tienda2,
        promedio_tienda3,
        promedio_tienda4
    ]
})

promedios
```

# 4. Productos más y menos vendidos

```python
max_vendido1 = tienda["Producto"].value_counts().idxmax()
max_vendido1
menos_vendido1 = tienda["Producto"].value_counts().idxmin()
menos_vendido1

max_vendido2 = tienda2["Producto"].value_counts().idxmax()
max_vendido2
menos_vendido2 = tienda2["Producto"].value_counts().idxmin()
menos_vendido2

max_vendido3 = tienda3["Producto"].value_counts().idxmax()
max_vendido3
menos_vendido3 = tienda3["Producto"].value_counts().idxmin()
menos_vendido3

max_vendido4 = tienda4["Producto"].value_counts().idxmax()
max_vendido4
menos_vendido4 = tienda4["Producto"].value_counts().idxmin()
menos_vendido4
```

```python
max_df = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Producto más vendido": [
        max_vendido1,
        max_vendido2,
        max_vendido3,
        max_vendido4
    ]})
max_df
```

```python
min_df = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Producto menos vendido": [
        menos_vendido1,
        menos_vendido2,
        menos_vendido3,
        menos_vendido4
    ]})
min_df
```
